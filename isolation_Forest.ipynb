```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.ensemble import IsolationForest
#generate synthetic data
np.random.seed=42
normal_data=np.random.randn(1000,2)*2 #normal data points
anomaly_data=np.random.randn(50,2)*10  #anomaly points
data=np.vstack([normal_data,anomaly_data]) #combine anomaly and normal
#fit isolation forest model
model=IsolationForest(contamination=0.05)
#expected proportion of anomalies
model.fit(data)
#predict anomalies
predictions=model.predict(data)
plt.scatter(data[:,0],data[:,1],c=predictions,cmap=plt.cm.Paired)
plt.title("Isolation Forest Anomaly Detection")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.show()
```

Write a program to detect anomalies in make_blobs dataset using Isolation Forest

```python

import numpy as np
import matplotlib.pyplot as plt
from sklearn.ensemble import IsolationForest
from sklearn.datasets import make_blobs

# Generate synthetic data using make_blobs
centers = [[0, 0], [5, 5], [10, 10]]
normal_data, _ = make_blobs(n_samples=1000, centers=centers, cluster_std=1.0, random_state=42)

# Generate some anomalies
anomaly_data = np.random.uniform(low=-10, high=20, size=(50, 2))

# Combine normal and anomaly data
data = np.vstack((normal_data, anomaly_data))

# Fit Isolation Forest model
model = IsolationForest(contamination=0.05, random_state=42)

# Contamination parameter sets expected proportion of anomalies in the data
model.fit(data)

# Predict anomalies
predictions = model.predict(data)

# Plot the data and detected anomalies
plt.scatter(data[:, 0], data[:, 1], c=predictions, cmap=plt.cm.Paired)
plt.title("Isolation Forest Anomaly Detection")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.show()
```

Write a program to detect anomalies in iris dataset using Isolation Forest

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.ensemble import IsolationForest
from sklearn.datasets import load_iris
from sklearn.preprocessing import StandardScaler

# Load the Iris dataset
iris = load_iris()
data = iris.data

# Standardize the data
scaler = StandardScaler()
data_scaled = scaler.fit_transform(data)

# Fit Isolation Forest model
model = IsolationForest(contamination=0.05, random_state=42)

# Contamination parameter sets expected proportion of anomalies in the data
model.fit(data_scaled)

# Predict anomalies
predictions = model.predict(data_scaled)

# Plot the data and detected anomalies (using first two features for 2D plot)
plt.scatter(data_scaled[:, 0], data_scaled[:, 1], c=predictions, cmap=plt.cm.Paired)
plt.title("Isolation Forest Anomaly Detection on Iris Dataset")
plt.xlabel("Feature 1 (Sepal Length)")
plt.ylabel("Feature 2 (Sepal Width)")
plt.show()
```
