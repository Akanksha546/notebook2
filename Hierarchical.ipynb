Hierarchical Clustering and plotting Dendrogram

```python
import numpy as np
from scipy.cluster.hierarchy import dendrogram, linkage
from matplotlib import pyplot as plt

# Distance matrix
X = np.array([[0, 0.71, 5.66, 3.61, 4.24,3.20],
              [0.71, 0, 4.95, 2.92,3.54,2.50],
              [5.66, 4.95, 0, 2.24,1.41,2.50],
              [3.61, 2.92, 2.24, 0,1.00,0.50],
              [4.24,3.54,1.41,1.00,0,1.12,],
              [3.20,2.50,2.50,0.50,1.12,0]])

# Perform hierarchical clustering
Z = linkage(X)

# Plot dendrogram
plt.figure(figsize=(10, 5))
dendrogram(Z)
plt.title('Hierarchical Clustering Dendrogram')
plt.xlabel('Data Points')
plt.ylabel('Distance')
plt.show()
```
